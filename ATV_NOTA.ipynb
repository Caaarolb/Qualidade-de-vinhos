```python
from ucimlrepo import fetch_ucirepo

from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
%matplotlib inline
```

# Identificando a tabela escolhida

```python
wine_quality = fetch_ucirepo(id=186)
```

```python
X = wine_quality.data.features
y = wine_quality.data.targets
```

```python
df = pd.concat([X, y], axis=1)
```

# 1 - Carregar os dados em uma tabela

#Carregando a tabela

```python
print
display(df.head())
```

# 2 - Fazer um gráfico de pontos de duas variáveis

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='alcohol', y='quality', color='darkred')
plt.title('Relação entre Álcool e Qualidade do Vinho')
plt.xlabel('Teor Alcoólico (%)')
plt.ylabel('Qualidade')
plt.show()
```

# 3 - Histograma de 4 variáveis com 10 e com 40 divisões.

```python
variaveis = ['alcohol', 'density', 'pH', 'citric_acid']
```

# 10 variáveis

```python
df[variaveis].hist(bins=10, figsize=(10,8), color='salmon', edgecolor='black')
plt.suptitle('Histogramas com 10 Divisões', fontsize=14)
plt.show()
```

# 40 variáveis

```python
df[variaveis].hist(bins=40, figsize=(10,8), color='lightcoral', edgecolor='black')
plt.suptitle('Histogramas com 40 Divisões', fontsize=14)
plt.show()
```

# 4 - Calcular média, mediana, moda e quartis para todas as variáveis.

```python
estatisticas = pd.DataFrame({
    'Média': df.mean(),
    'Mediana': df.median(),
    'Moda': df.mode().iloc[0],
    '1º Quartil (Q1)': df.quantile(0.25),
    '3º Quartil (Q3)': df.quantile(0.75)
})
```

# Estastisticas dos resultados

```python
print
display(estatisticas.round(3))
```
